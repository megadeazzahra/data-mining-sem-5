## Membersihkan Data

### hapus baris bernilai NaN

```python
df = df.dropna(subset=['gross', 'budget'])
```

Menghapus Baris yang Hilang: Kita menggunakan df.dropna(subset=['gross', 'budget']) untuk menghapus semua baris yang memiliki nilai NaN pada kolom 'gross' dan 'budget'.

```python
df.isnull().sum()
```

Setelah menghapus baris, kita melakukan pemeriksaan ulang dengan df.isnull().sum() untuk memastikan bahwa tidak ada nilai hilang pada kolom yang ditargetkan.

### normalisasi kolom color

```python
# Mengubah kolom 'color' menjadi lowercase
df['color'] = df['color'].str.lower()

print(df)
```

Kita mengubah semua nilai dalam kolom 'color' menjadi huruf kecil dengan menggunakan metode .str.lower(). Normalisasi ini membantu dalam menjaga konsistensi data, sehingga ketika kita melakukan analisis atau pencarian, tidak ada perbedaan akibat penggunaan huruf kapital atau kecil.

### hapus nilai yang tidak standar

```python
df.isna()
```

```python
df.dropna(inplace=True)
print (df.isnull().sum())
```

Dengan df.dropna(inplace=True), kita menghapus semua baris yang masih memiliki nilai NaN di DataFrame. Kita melakukan pemeriksaan ulang dengan print(df.isnull().sum()) untuk memastikan bahwa tidak ada nilai hilang yang tersisa setelah proses pembersihan.

## Transformasi Data

### ubah tipe data

```python
df[['gross', 'budget']] = df[['gross', 'budget']].apply(pd.to_numeric, errors='coerce')
print (df.dtypes)
```

Kita menggunakan df[['gross', 'budget']].apply(pd.to_numeric, errors='coerce') untuk mengonversi kolom 'gross' dan 'budget' menjadi tipe data numerik. Dengan pengaturan errors='coerce', nilai yang tidak dapat diubah akan diubah menjadi NaN.

### memisahkan kolom genre

```python
print(df['genres'].head())
```

```python
# pisahkan genre
genres_split = df['genres'].str.split('|', expand=True)
```

Dengan df['genres'].str.split('|', expand=True), kita memisahkan nilai-nilai di kolom 'genres' yang terpisah oleh tanda '|'. Hal ini menghasilkan DataFrame baru di mana setiap genre ditempatkan dalam kolom terpisah.

```python
print(genres_split.head())
```

```python
# Beri nama kolom baru
genres_split.columns = [f'genre_{i+1}' for i in range(genres_split.shape[1])]

# Gabungkan dengan DataFrame asli
df = pd.concat([df, genres_split], axis=1)
```

Kita memberikan nama pada kolom baru dengan format genre_1, genre_2, dan seterusnya, berdasarkan jumlah kolom hasil pemisahan. Kita menggunakan pd.concat([df, genres_split], axis=1) untuk menggabungkan DataFrame asli dengan DataFrame genre yang baru dibuat secara horizontal.



```python
df.drop(columns=['genres'], inplace=True)
```

dengan df.drop(columns=['genres'], inplace=True), kita menghapus kolom 'genres' yang sudah tidak diperlukan lagi setelah pemisahan.

```python
# tampilkan dataframe
df.head()
```

### menormalisasi teks lowercase

```python
# Menormalisasi kolom teks menjadi huruf kecil
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

# Menampilkan DataFrame yang sudah dinormalisasi
df.head()
```

Kita menggunakan df.apply(lambda x: x.str.lower() if x.dtype == "object" else x) untuk mengubah semua kolom yang berisi teks menjadi huruf kecil.
